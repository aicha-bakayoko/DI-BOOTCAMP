```python
# =====================================================================
# SIRA LABS - SEMAINE 6 : LLM & IA GÉNÉRATIVE
# NOTEBOOK DE COMPLÉTION DES EXERCICES XP
# =====================================================================

# ---------------------------------------------------------------------
# 1. INSTALLATION ET CHARGEMENT DES MODÈLES (Exercice 1)
# ---------------------------------------------------------------------
print("1. Installation des dépendances et chargement du modèle...")
!pip install transformers matplotlib torch --quiet

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Choix du modèle causal GPT-2
model_name = "gpt2" 

# Chargement du tokenizer et du modèle pré-entraîné
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print(f"\nModel '{model_name}' loaded successfully!")
print("""
GPT-2 is a causal language model, meaning it predicts the next word in a sequence. 
It has been trained on a diverse dataset and can generate coherent, contextually relevant text.
""")

# ---------------------------------------------------------------------
# 2. PROCESSUS DE TOKENISATION ET VISUALISATION (Exercice 2 & 3)
# ---------------------------------------------------------------------
print("\n2. Exécution du processus de tokenisation...")

# Définition du texte à analyser
text = "Artificial Intelligence is transforming our world"

# Découpage en tokens et conversion en identifiants uniques (IDs)
tokens = tokenizer.tokenize(text) 
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print(f"Original Text : {text}")
print(f"Tokens        : {tokens}")
print(f"Token IDs     : {token_ids}\n")

# Affichage détaillé de la correspondance Jeton -> ID (Exercice 3.1)
print("Correspondance détaillée :")
for t, _id in zip(tokens, token_ids):
    print(f"  Token: {t:<12} -> ID: {_id}")

# Génération du graphique en barres (Exercice 2.3)
plt.figure(figsize=(10, 4))
plt.bar(tokens, token_ids, color="skyblue")
plt.xlabel("Tokens")
plt.ylabel("Token IDs")
plt.title("Visualisation des Token IDs par Jeton (GPT-2)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------------------------------------------------------------------
# 3. GÉNÉRATION AUTOMATIQUE DE TEXTE (Exercice 5)
# ---------------------------------------------------------------------
print("\n3. Initialisation du pipeline de génération de texte...")

# Définition d'une phrase d'amorce (Prompt)
input_text = "The future of AI in Africa is"

# Création du pipeline Hugging Face pour l'inférence
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Génération d'une suite logique de 30 mots maximum
output = generator(input_text, max_length=30, num_return_sequences=1)
output_text = output[0]['generated_text']

print("\n--- RÉSULTAT DE LA GÉNÉRATION ---")
print(f"Input  : {input_text}")
print(f"Output : {output_text}")
print("---------------------------------")
```
