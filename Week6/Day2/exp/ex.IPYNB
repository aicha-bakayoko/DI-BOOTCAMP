```python
{
  "cells": [
    {
      "cell_type": "markdown",
      "metadata": {
        "id": "cell-1"
      },
      "source": [
        "# Week 6 - Day 2: Deep Dive into Advanced Transformer and LLM Concepts\n",
        "**Student:** N'guessan Lylda Rachelle Yao  \n",
        "**Cohort:** COT_GenAI & Machine Learning Bootcamp - 2026\n",
        "\n",
        "---"
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {
        "id": "cell-2"
      },
      "source": [
        "## 🌟 Exercise 1: Traditional vs. Modern NLP: A Comparative Analysis\n",
        "\n",
        "### 1. Comparison Table\n",
        "\n",
        "| Feature | Traditional NLP Paradigm | Modern NLP Paradigm (LLMs) |\n",
        "| :--- | :--- | :--- |\n",
        "| **Feature Engineering** | **Manual:** Requiert une expertise humaine pour concevoir des caractéristiques (Regex, POS tagging, TF-IDF). | **Automatic:** Le modèle apprend directement les représentations utiles à partir des données brutes. |\n",
        "| **Word Representations**| **Static:** Un mot a le même vecteur peu importe le contexte (ex: *Word2Vec*, *GloVe* où \"avocat\" [fruit] = \"avocat\" [métier]). | **Contextual:** La représentation d'un mot change dynamiquement selon les mots qui l'entourent (ex: *BERT*, *GPT*). |\n",
        "| **Model Architectures** | **Shallow:** Modèles statistiques ou neuronaux simples (Naïve Bayes, SVM, RNN/LSTM superficiels). | **Deep:** Architectures Transformers massives contenant des milliards de paramètres. |\n",
        "| **Training Methodology**| **Task-specific:** Entraîné de zéro (from scratch) pour une tâche unique avec un jeu de données labellisé dédié. | **Pre-training / Fine-tuning:** Pré-entraînement auto-supervisé sur corpus géant, puis ajustement fin sur la tâche cible. |\n",
        "| **Key Examples** | Naïve Bayes, SVM, TF-IDF + Logistic Regression, early LSTMs. | BERT, GPT-4, RoBERTa, T5, LLaMA. |\n",
        "| **Advantages** | * Légers, rapides à entraîner.<br>* Interprétables.<br>* Exigent peu de ressources de calcul (CPU suffit). | * Performances de pointe (SoTA).<br>* Flexibles (un seul modèle pour plusieurs tâches).<br>* Moins besoin de données labellisées. |\n",
        "| **Disadvantages** | * Ne capturent pas le contexte long terme.<br>* Manquent de généralisation.<br>* Coûteux en ingénierie de caractéristiques. | * Très gourmands en ressources (GPU/TPU).<br>* Effet \"boîte noire\" (difficile à interpréter).<br>* Risques d'hallucinations et de biais. |\n",
        "\n",
        "### 2. Impact on Scalability and Efficiency\n",
        "L'évolution vers le NLP moderne a radicalement transformé l'évolutivité (scalability) et l'efficacité :\n",
        "* **Scalabilité du calcul :** L'architecture Transformer élimine la nature séquentielle des RNN/LSTM. Grâce au mécanisme d'attention, le traitement s'effectue en parallèle, permettant d'entraîner des modèles sur des jeux de données de taille Internet (pétaoctets) en exploitant pleinement les clusters de GPU.\n",
        "* **Efficacité du transfert (Transfer Learning) :** Auparavant, créer une application NLP exigeait de collecter des milliers de données annotées pour chaque nouvelle tâche. Aujourd'hui, l'approche *Pre-training/Fine-tuning* réutilise les connaissances générales d'un modèle global. Cela réduit le besoin en données labellisées et le temps de convergence pour les tâches spécifiques, rendant le déploiement industriel beaucoup plus rapide et rentable."
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {
        "id": "cell-3"
      },
      "source": [
        "## 🌟 Exercise 2: LLM Architecture and Application Scenarios\n",
        "\n",
        "### 1. BERT (Bidirectional Encoder Representations from Transformers)\n",
        "* **Core Architecture:** **Encoder-only**. Il utilise un mécanisme d'attention **bidirectionnel**, ce qui signifie qu'il regarde simultanément à gauche et à droite d'un mot pour en comprendre le sens. Son objectif de pré-entraînement principal est le *Masked Language Modeling* (MLM).\n",
        "* **Real-World Application:** Systèmes de recherche textuelle et extraction d'informations (ex: Google Search algorithm).\n",
        "* **Suitability Justification:** Pour comprendre l'intention derrière une requête de recherche (\"vol Paris New York sans escale\"), le modèle doit impérativement analyser la phrase entière de manière globale. La bidirectionnalité de BERT lui permet de saisir parfaitement les nuances contextuelles et les relations subtiles entre les mots.\n",
        "\n",
        "### 2. GPT (Generative Pre-trained Transformer)\n",
        "* **Core Architecture:** **Decoder-only**. Il utilise un mécanisme d'attention **unidirectionnel (masqué)**. Le modèle ne peut regarder que les tokens précédents (à gauche) pour prédire le mot suivant. Son objectif est le *Causal Language Modeling* (CLM).\n",
        "* **Real-World Application:** Rédaction assistée, génération de code et agents conversationnels (ex: ChatGPT, GitHub Copilot).\n",
        "* **Suitability Justification:** La génération de texte est un processus intrinsèquement auto-régressif (mot après mot). L'architecture causale de GPT est parfaitement optimisée pour maintenir la cohérence fluide du texte généré au fur et à mesure de sa progression.\n",
        "\n",
        "### 3. T5 (Text-to-Text Transfer Transformer)\n",
        "* **Core Architecture:** **Encoder-Decoder**. Il combine un encodeur bidirectionnel standard et un décodeur causal. Il reformule chaque tâche NLP sous la forme d'un problème \"Text-to-Text\" (Entrée: texte $\\rightarrow$ Sortie: texte).\n",
        "* **Real-World Application:** Résumé de documents longs (Abstractive Summarization) ou traduction automatique.\n",
        "* **Suitability Justification:** Le résumé requiert d'abord une compréhension profonde et globale du texte source (rôle de l'encodeur bidirectionnel), suivie de la génération d'un tout nouveau texte condensé et fluide (rôle du décodeur). L'architecture hybride de T5 excelle dans cette transition d'un espace de sens à un autre."
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {
        "id": "cell-4"
      },
      "source": [
        "## 🌟 Exercise 3: The Benefits and Ethical Considerations of Pre-training\n",
        "\n",
        "### 1. Five Key Benefits of Pre-trained Models\n",
        "* **Improved Generalization:** En s'entraînant sur des textes massifs et variés, le modèle développe une compréhension globale de la grammaire, de la syntaxe et des concepts du monde réel, lui permettant de bien réagir face à des données qu'il n'a jamais vues.\n",
        "* **Reduced Need for Labeled Data:** Le pré-entraînement est auto-supervisé (sans étiquettes humaines). Lors du fine-tuning, le modèle a seulement besoin d'une infime fraction de données labellisées pour atteindre d'excellentes performances.\n",
        "* **Faster Fine-tuning:** Le modèle possédant déjà les bases linguistiques, l'ajustement pour une tâche spécifique ne demande que quelques époques d'entraînement, économisant du temps et de l'énergie.\n",
        "* **Transfer Learning:** C'est la capacité de transférer les connaissances acquises sur un problème général (prédire des mots manquants) vers un problème spécialisé (analyser des contrats juridiques).\n",
        "* **Robustness:** Les modèles pré-entraînés gèrent mieux les ambiguïtés, les fautes de frappe ou les variations de style linguistique grâce à la richesse statistique de leur entraînement initial.\n",
        "\n",
        "### 2. Ethical Concerns\n",
        "* **Bias:** Les jeux de données du web contiennent des stéréotypes de genre, de race, de religion et de classe. Les LLMs absorbent ces structures et peuvent propager ou amplifier ces discriminations dans leurs réponses.\n",
        "* **Misinformation:** Les LLMs génèrent des affirmations fausses mais présentées de manière très convaincante (hallucinations), ce qui peut servir à saturer le web de fake news à grande échelle.\n",
        "* **Misuse:** Utilisation malveillante pour générer du code pour des malwares, automatiser des campagnes de phishing personnalisé ou tricher aux examens académiques.\n",
        "\n",
        "### 3. Mitigation Strategies\n",
        "* **Curated & Audited Datasets:** Filtrer activement les données de pré-entraînement pour supprimer les discours de haine, la toxicité et les données personnelles (PII).\n",
        "* **RLHF (Reinforcement Learning from Human Feedback):** Aligner le modèle après son entraînement en utilisant des retours humains pour le pénaliser s'il génère des contenus toxiques ou faux.\n",
        "* **Guardrails & Content Filtering:** Mettre en place des filtres de sécurité stricts en entrée (prompts) et en sortie (réponses) pour bloquer les requêtes dangereuses ou illégales."
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {
        "id": "cell-5"
      },
      "source": [
        "## 🌟 Exercise 4: Transformer Architecture Deep Dive\n",
        "\n",
        "### 1. Self-Attention and Multi-Head Attention\n",
        "* **Self-Attention Mechanism:** Pour chaque mot (token), le modèle génère trois vecteurs : une Requête (**Query** $Q$), une Clé (**Key** $K$) et une Valeur (**Value** $V$). Le modèle calcule un score d'attention en faisant le produit scalaire entre le vecteur $Q$ d'un mot et les vecteurs $K$ de tous les autres mots de la phrase. Ces scores sont normalisés via une fonction Softmax, puis utilisés pour faire une moyenne pondérée des vecteurs $V$. Cela permet au mot de \"zoomer\" sur les éléments les plus importants de son contexte.\n",
        "* **Multi-Head Attention Purpose:** Au lieu de calculer l'attention une seule fois, le modèle sépare les vecteurs $Q, K, V$ en plusieurs sous-espaces (les \"têtes\"). Chaque tête peut se concentrer sur un type de relation linguistique différent (relations syntaxiques, pronoms, temps des verbes) en même temps. \n",
        "* **Concrete Example:** Prenons la phrase : *\"L'ingénieur a réparé la machine car elle était en panne.\"*\n",
        "  * **Tête d'attention 1** va lier le pronom **\"elle\"** directement au nom **\"machine\"** (résolution de coréférence).\n",
        "  * **Tête d'attention 2** va lier le verbe **\"réparé\"** au sujet **\"ingénieur\"** (relation sujet-verbe).\n",
        "\n",
        "### 2. Pre-training Objectives\n",
        "* **Comparison:**\n",
        "  * **MLM (Masked Language Modeling) :** Le modèle masque aléatoirement certains mots de la phrase (ex: 15%) et doit deviner ces mots en utilisant le contexte gauche **et** droite (bidirectionnel).\n",
        "  * **CLM (Causal Language Modeling) :** Le modèle doit prédire le mot suivant en se basant uniquement sur la séquence des mots passés (unidirectionnel).\n",
        "* **Scenarios:**\n",
        "  * **MLM scenario:** Analyse de sentiments ou classification de textes médicaux, où capter le contexte global exact est crucial.\n",
        "  * **CLM scenario:** Écriture d'un assistant de code ou d'un agent de support client interactif (génération de texte fluide).\n",
        "* **Next Sentence Prediction (NSP):** Au début, BERT utilisait l'NSP (prédire si la phrase B suit la phrase A) pour aider le modèle à comprendre les relations inter-phrases. Cependant, des recherches ultérieures (comme RoBERTa) ont montré que cette tâche était trop simple et dégradait parfois les performances. Entraîner le modèle sur de plus longues séquences continues s'est avéré bien plus efficace.\n",
        "\n",
        "### 3. Transformer Model Selection\n",
        "* **A. Sentiment Analysis System:**\n",
        "  * **Type:** **Encoder-only** (ex: BERT/RoBERTa).\n",
        "  * **Justification:** La classification de texte nécessite une compréhension fine et globale de l'intégralité du texte pour repérer le sarcasme ou les nuances. L'attention bidirectionnelle permet d'encoder efficacement l'ensemble du retour client en un seul vecteur de classification.\n",
        "* **B. Creative Chatbot:**\n",
        "  * **Type:** **Decoder-only** (ex: GPT-4, LLaMA).\n",
        "  * **Justification:** Un chatbot doit générer des réponses mot après mot de manière fluide et dynamique. L'attention causale (unidirectionnelle) empêche le modèle de tricher en regardant vers l'avenir, ce qui est parfait pour la génération ouverte.\n",
        "* **C. Document Translation Service:**\n",
        "  * **Type:** **Encoder-Decoder** (ex: T5, BART).\n",
        "  * **Justification:** La traduction nécessite d'abord d'ingérer l'intégralité de la phrase source en anglais sans contrainte directionnelle (Encodeur), puis de mapper ce sens pour générer la phrase cible en espagnol de manière fluide.\n",
        "\n",
        "### 4. Positional Encoding\n",
        "* **Purpose:** Comme l'architecture Transformer traite tous les mots d'une phrase simultanément (parallélisme), elle n'a aucune notion innée de l'ordre des mots (contrairement aux RNN). Le *Positional Encoding* ajoute un vecteur mathématique aux embeddings de mots pour injecter l'information de leur position/ordre dans la phrase.\n",
        "* **Problem Example:** Sans encodage positionnel, le Transformer verrait la phrase *\"Le chat a mangé la souris\"* et *\"La souris a mangé le chat\"* comme étant strictement **identiques**, car elles contiennent exactement le même sac de mots, ce qui fausserait complètement le sens."
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {
        "id": "cell-6"
      },
      "source": [
        "## 🌟 Exercise 5: BERT Variations - Choose Your Detective\n",
        "\n",
        "### 1. Scenario Identification\n",
        "* **Scenario 1:** *Real-time sentiment analysis on mobile app* $\\rightarrow$ **DistilBERT**. Il est compact, rapide et conçu spécifiquement pour s'exécuter avec de faibles ressources tout en conservant 97% des capacités de BERT.\n",
        "* **Scenario 2:** *Research on legal documents requiring high accuracy* $\\rightarrow$ **RoBERTa**. Entraîné plus longtemps, sur des volumes de données beaucoup plus vastes et sans la tâche restrictive NSP, il offre une précision maximale.\n",
        "* **Scenario 3:** *Global customer support in multiple languages* $\\rightarrow$ **XLM-RoBERTa**. Il est pré-entraîné sur des textes multilingues massifs (plus de 100 langues), idéal pour traiter des requêtes internationales.\n",
        "* **Scenario 4:** *Efficient pretraining, and token replacement detection* $\\rightarrow$ **ELECTRA**. Il remplace le MLM traditionnel par une tâche de détection de tokens corrompus générés par un petit réseau de neurones, rendant l'apprentissage beaucoup plus efficace par échantillon.\n",
        "* **Scenario 5:** *Efficient NLP in resource-constrained environments* $\\rightarrow$ **ALBERT**. Il réduit drastiquement le nombre de paramètres de BERT grâce au partage des paramètres entre les couches et à la décomposition des matrices d'embedding, réduisant l'empreinte mémoire.\n",
        "\n",
        "### 2. BERT Variations Comparison Table\n",
        "\n",
        "| Model | Key Optimization / Innovation | Size & Efficiency | Ideal Use Case |\n",
        "| :--- | :--- | :--- | :--- |\n",
        "| **RoBERTa** | Supprime l'NSP, utilise le masquage dynamique et un entraînement sur de plus gros batches avec plus de données. | Taille similaire à BERT mais nécessite de grosses ressources pour l'entraînement. | Tâches de haute précision (Recherche académique, juridique). |\n",
        "| **ALBERT** | Partage des paramètres inter-couches et factorisation de la matrice d'embedding des tokens. | Très petite taille mémoire (fichiers légers), calculs modérés. | Serveurs légers ou applications exigeant une faible empreinte RAM. |\n",
        "| **DistilBERT**| Utilise la distillation de connaissances (Knowledge Distillation) d'un grand BERT vers un petit modèle. | 40% plus petit, 60% plus rapide que BERT. | Applications mobiles et APIs temps réel à faible latence. |\n",
        "| **ELECTRA** | Remplace le masquage par une tâche de \"Replaced Token Detection\" (type GAN/Discriminateur). | Très efficace : atteint de hautes performances avec moins de temps de calcul. | Pré-entraînement rapide avec un budget informatique limité. |\n",
        "| **XLM-RoBERTa**| Entraînement multi-langues basé sur l'architecture robuste de RoBERTa sur le dataset CommonCrawl. | Volumineux en raison d'un vocabulaire étendu (plusieurs langues). | Applications cross-lingues et support client multilingue. |"
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {
        "id": "cell-7"
      },
      "source": [
        "## 🌟 Exercise 6: Softmax Temperature - The Randomness Regulator\n",
        "\n",
        "### 1. Temperature Scenarios\n",
        "* **Temperature = 0.2 (Basse) :** La distribution de probabilité devient très pointue. Le modèle choisit presque toujours le token le plus probable. Les réponses seront hautement prévisibles, répétitives, cohérentes mais sans aucune originalité.\n",
        "* **Temperature = 1.0 (Standard) :** Le modèle échantillonne les mots en suivant fidèlement les probabilités originales apprises durant son entraînement. C'est l'équilibre parfait entre créativité naturelle et structure grammaticale.\n",
        "* **Temperature = 1.5 (Haute) :** La distribution s'aplatit (les mots improbables reçoivent une chance élevée d'être choisis). Le texte devient très chaotique, créatif, mais va rapidement perdre sa cohérence logique et générer des fautes de syntaxe ou du non-sens.\n",
        "\n",
        "### 2. Application Design\n",
        "* **Bedtime Stories for Children:** Je configurerais la température autour de **0.7 ou 0.8**. Cela donne assez de liberté au modèle pour inventer des rebondissements magiques et originaux (créativité), tout en restant assez bas pour s'assurer que l'histoire garde un fil conducteur logique et une grammaire correcte pour un enfant.\n",
        "* **Financial Reports Summaries:** Je configurerais la température à **0.0 (ou proche de 0.1)**. Dans le domaine de la finance, l'imagination est un défaut grave. On veut une extraction pure, factuelle et déterministe des chiffres et des données du rapport, sans aucune fioriture ni hallucination.\n",
        "\n",
        "### 3. Temperature and Bias\n",
        "* **Impact of Temperature on Bias:** Une température **basse** (ex: 0.1) force le modèle à sortir les stéréotypes les plus fréquents ancrés dans ses données d'entraînement (biais majoritaire). Une température **haute** (ex: 1.2) réduit la probabilité des stéréotypes évidents, mais augmente le risque de générer du contenu toxique ou des insultes aléatoires si le modèle commence à divaguer.\n",
        "* **Practical Example:** Si on donne le prompt *\"Le chirurgien est entré dans la salle et...\"* :\n",
        "  * À **T = 0.1**, le modèle choisira de manière déterministe le pronom masculin *\"il\"* en raison du biais de genre historique présent dans les textes de son dataset.\n",
        "  * À **T = 0.9**, la distribution s'élargit, laissant au modèle une probabilité accrue de sélectionner *\"elle\"* ou de reformuler la phrase de manière neutre."
      ]
    }
  ],
  "metadata": {
    "language_info": {
      "name": "python"
    }
  },
  "nbformat": 4,
  "nbformat_minor": 2
}
```
