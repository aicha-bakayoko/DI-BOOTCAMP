```python
# =====================================================================
# SYSTEM COMPONENT: IMPORTATION DES LIBRAIRIES (Exercice 3)
# =====================================================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Configuration des graphiques
plt.style.use('default')
print("✅ Librairies importées avec succès.")


# =====================================================================
# EXERCICE 1 : Deep Learning vs. Traditional Machine Learning
# =====================================================================
"""
--- QUESTION 1 : Tableau Comparatif ---
| Aspect | Machine Learning Traditionnel | Deep Learning |
| :--- | :--- | :--- |
| Feature Engineering | Manuel, nécessite l'intervention d'experts métiers. | Automatique, le modèle apprend les features par lui-même. |
| Data Processing | Fonctionne bien avec des données tabulaires/structurées. | Conçu pour les données massives et non structurées. |
| Scalability | Performance plafonne même avec plus de données. | Performance s'améliore continuellement avec plus de données. |
| Pattern Discovery | Limité aux relations linéaires ou simples. | Découvre des motifs complexes et non linéaires profonds. |
| Computational Requirements| Faible à modéré (CPU standard). | Très élevé (nécessite des GPU/TPU). |

--- QUESTION 2 : Cas d'usage réels ---
- Machine Learning Traditionnel : Prédire le prix d'une maison basé sur des critères tabulaires (superficie, quartier).
- Deep Learning : Détection de tumeurs sur des images de scanners médicaux (IRM).

--- QUESTION 3 : Explication (Unstructured Data) ---
Le Deep Learning possède un avantage majeur pour les données non structurées (images, audio, texte) car il élimine le besoin d'extraire manuellement les caractéristiques. Grâce aux couches successives de neurones, le modèle apprend lui-même une hiérarchie de caractéristiques complexes directement à partir des données brutes. Par exemple, il détecte d'abord des bords, puis des formes, puis des objets entiers, ce qui est impossible à coder efficacement à la main avec du ML traditionnel.
"""
print("\n🔹 Exercice 1 complété (Réponses dans les commentaires du code).")


# =====================================================================
# EXERCICE 2 : Artificial Neural Networks (ANNs)
# =====================================================================
"""
--- QUESTION 1 & 2 : Structure du Réseau ---
[Architecture représentée] :
- Couche d'entrée (Input Layer) : 3 neurones (représentant les variables d'entrée)
- Couche cachée (Hidden Layer) : 4 neurones connectés aux entrées avec fonction d'activation
- Couche de sortie (Output Layer) : 2 neurones (pour les prédictions ou la classification)
- Composants clés : Neurones (unités de calcul), Poids (Weights / force de la connexion), 
  Biais (Biases / ajustement linéaire), Fonction d'activation (introduction de la non-linéarité).

--- QUESTION 3 : Flux d'information ---
L'information traverse le réseau de manière unidirectionnelle de l'entrée vers la sortie (Forward Propagation). 
Les données de la couche d'entrée sont multipliées par des poids, combinées à un biais, puis passées à travers 
une fonction d'activation au sein des neurones de la couche cachée. Ce processus de combinaison linéaire et 
de transformation non linéaire se répète de couche en couche jusqu'à ce que la couche de sortie génère la prédiction finale.
"""
print("🔹 Exercice 2 complété (Réponses dans les commentaires du code).")


# =====================================================================
# EXERCICE 3 : Création du Dataset et Visualizing the Data
# =====================================================================
print("\n🔹 Exécution de l'Exercice 3...")

# 1 & 2. Génération des données avec bruit
np.random.seed(0)
x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))

# 4. Split Train (12 points) / Test (8 points)
x_train, x_test = x[:12], x[12:]
y_train, y_test = y[:12], y[12:]

# 3. Graphique de dispersion (Scatter plot)
plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color='blue', label='Train Set (12 points)', s=50)
plt.scatter(x_test, y_test, color='red', label='Test Set (8 points)', s=50)
plt.title('Exercice 3 : Visualisation du Dataset avec Bruit ($y = -x^2$)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()


# =====================================================================
# EXERCICE 4 : Fitting Polynomial Models of Different Degrees
# =====================================================================
print("\n🔹 Exécution de l'Exercice 4...")

# 1. Fonction d'ajustement polynomial
def polynomial_fit(degree):
    # Calcule les coefficients optimaux
    coefficients = np.polyfit(x_train, y_train, degree)
    # Renvoie une fonction utilisable pour les prédictions
    return np.poly1d(coefficients)

# 2. Fonction d'affichage des courbes
def plot_polyfit(degree):
    model = polynomial_fit(degree)
    
    # Abscisse continue pour un tracé fluide de la courbe
    x_curve = np.linspace(-1, 1, 200)
    y_curve = model(x_curve)
    
    plt.scatter(x_train, y_train, color='blue', label='Train Set', s=40)
    plt.scatter(x_test, y_test, color='red', label='Test Set', s=40)
    plt.plot(x_curve, y_curve, label=f'Degré {degree}', color='green', linewidth=2)
    plt.title(f'Ajustement Degré {degree}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim(-1.5, 0.5)  # Évite les distorsions visuelles extrêmes de l'overfitting
    plt.legend()

# 3 & 4. Visualisation pour les degrés 1, 7 et 11
degrees_to_plot = [1, 7, 11]
plt.figure(figsize=(16, 4))

for i, deg in enumerate(degrees_to_plot, 1):
    plt.subplot(1, 3, i)
    plot_polyfit(deg)

plt.tight_layout()
plt.show()

print("""
--- Observations de l'Exercice 4 ---
- Degré 1 : Le modèle est une ligne droite. Il y a sous-apprentissage (Underfitting) car il ne capte pas la courbure naturelle des données.
- Degré 7 : Le modèle suit bien la forme générale, mais commence à osciller pour capturer le bruit des points d'entraînement.
- Degré 11 : Le modèle passe par absolument tous les points bleus (Train). C'est un surapprentissage (Overfitting) total. La courbe fait n'importe quoi aux extrémités et échoue complètement à prédire le set de Test (points rouges).
""")


# =====================================================================
# EXERCICE 5 : Cross-Validation to Find the Optimal Degree
# =====================================================================
print("\n🔹 Exécution de l'Exercice 5...")

results = []

# 1 & 2 & 3. Boucle et calcul des scores RMSE pour chaque degré
for degree in range(1, 12):
    # Entraînement sur le train set
    model = np.poly1d(np.polyfit(x_train, y_train, degree))
    
    # Calcul des prédictions
    y_train_pred = model(x_train)
    y_test_pred = model(x_test)
    
    # Calcul de la racine de l'erreur quadratique moyenne (RMSE)
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    results.append((degree, rmse_train, rmse_test))

# Extraction pour le graphique
degrees = [r[0] for r in results]
rmse_train_vals = [r[1] for r in results]
rmse_test_vals = [r[2] for r in results]

# 4. Tracé avec échelle logarithmique sur l'axe Y
plt.figure(figsize=(8, 5))
plt.plot(degrees, rmse_train_vals, marker='o', label='RMSE Entraînement (Train)', color='blue')
plt.plot(degrees, rmse_test_vals, marker='o', label='RMSE Test', color='red')
plt.yscale('log')
plt.xlabel('Degré du Polynôme')
plt.ylabel('RMSE (Échelle Logarithmique)')
plt.title('Exercice 5 : Évolution du RMSE vs Degré du Polynôme')
plt.xticks(degrees)
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

# 5. Identification du degré optimal
optimal_idx = np.argmin(rmse_test_vals)
optimal_degree = degrees[optimal_idx]

print(f"--- Conclusion finale ---")
print(f"Le degré qui minimise l'erreur sur les données de test est le **Degré {optimal_degree}**.")
print(f"Cela confirme parfaitement le modèle réel sous-jacent qui était d'ordre 2 ($y = -x^2$).")
```
