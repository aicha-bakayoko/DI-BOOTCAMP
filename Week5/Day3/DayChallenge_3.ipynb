## Défi quotidien : Classification de chiffres manuscrits à l'aide de réseaux de neurones convolutifs (CNN)

Ce notebook vous guidera à travers la classification de chiffres manuscrits en utilisant deux types de réseaux de neurones : un réseau neuronal entièrement connecté (Dense) et un réseau neuronal convolutif (CNN). Nous utiliserons l'ensemble de données MNIST.

### 1. Charger l'ensemble de données MNIST

Nous allons charger l'ensemble de données MNIST en utilisant `keras.datasets.mnist.load_data()` et imprimer les formes des données pour comprendre leur structure.

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

# Charger l'ensemble de données MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Imprimer les formes des données chargées
print(f"Forme de x_train: {x_train.shape}")
print(f"Forme de y_train: {y_train.shape}")
print(f"Forme de x_test: {x_test.shape}")
print(f"Forme de y_test: {y_test.shape}")
```

### 2. Prétraitement des données pour un réseau neuronal entièrement connecté

Pour un réseau neuronal entièrement connecté, nous devons:
- Aplatir les images (de 28x28 à 784 pixels).
- Normaliser les valeurs des pixels en les divisant par 255.
- Encoder les étiquettes cibles en one-hot.

```python
# Aplatir les images
x_train_flat = x_train.reshape(-1, 28*28)
x_test_flat = x_test.reshape(-1, 28*28)

# Normaliser les valeurs des pixels
x_train_flat = x_train_flat / 255.0
x_test_flat = x_test_flat / 255.0

# Encoder les étiquettes cibles en one-hot
y_train_one_hot = keras.utils.to_categorical(y_train, num_classes=10)
y_test_one_hot = keras.utils.to_categorical(y_test, num_classes=10)

print(f"Forme de x_train_flat après aplatissement et normalisation: {x_train_flat.shape}")
print(f"Forme de y_train_one_hot après encodage one-hot: {y_train_one_hot.shape}")
```

### 3. Construire et entraîner un réseau neuronal entièrement connecté

Nous allons créer un modèle `Sequential` avec des couches `Dense`, le compiler et l'entraîner.

```python
model_dense = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(784,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Compiler le modèle
model_dense.compile(optimizer='adam',
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

# Entraîner le modèle
history_dense = model_dense.fit(x_train_flat, y_train_one_hot, epochs=5, validation_split=0.2)

# Évaluer les performances
loss_dense, accuracy_dense = model_dense.evaluate(x_test_flat, y_test_one_hot)
print(f"\nPrécision du modèle Dense sur l'ensemble de test: {accuracy_dense:.4f}")
```

### 4. Prétraitement des données pour un réseau neuronal convolutif

Pour un CNN, nous devons:
- Remodeler les données d'entrée pour ajouter une dimension de canal (e.g., (60000, 28, 28, 1)).
- Normaliser les valeurs des pixels en les divisant par 255.
- Encoder les étiquettes cibles en one-hot (déjà fait, mais nous utiliserons les mêmes `y_train_one_hot` et `y_test_one_hot`).

```python
# Remodeler les données d'entrée pour CNN
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)

# Normaliser les valeurs des pixels (déjà fait pour flat, mais on le refait pour la clarté)
x_train_cnn = x_train_cnn / 255.0
x_test_cnn = x_test_cnn / 255.0

print(f"Forme de x_train_cnn après remodelage et normalisation: {x_train_cnn.shape}")
```

### 5. Concevoir et entraîner un réseau neuronal convolutif

Nous allons créer un modèle `Sequential` avec des couches `Conv2D`, `MaxPool2D`, `Flatten`, et `Dense`, le compiler et l'entraîner.

```python
model_cnn = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Compiler le modèle
model_cnn.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Entraîner le modèle
history_cnn = model_cnn.fit(x_train_cnn, y_train_one_hot, epochs=5, validation_split=0.2)

# Évaluer les performances
loss_cnn, accuracy_cnn = model_cnn.evaluate(x_test_cnn, y_test_one_hot)
print(f"\nPrécision du modèle CNN sur l'ensemble de test: {accuracy_cnn:.4f}")
```

### 6. Comparer les performances

Comparons la précision des deux modèles et observons la différence.

```python
print("\n--- Comparaison des performances des modèles ---")
print(f"Précision du modèle Entièrement Connecté: {accuracy_dense:.4f}")
print(f"Précision du modèle CNN: {accuracy_cnn:.4f}")

if accuracy_cnn > accuracy_dense:
    print("Le modèle CNN a obtenu une meilleure précision que le modèle entièrement connecté.")
elif accuracy_cnn < accuracy_dense:
    print("Le modèle entièrement connecté a obtenu une meilleure précision que le modèle CNN.")
else:
    print("Les deux modèles ont obtenu une précision similaire.")
```
