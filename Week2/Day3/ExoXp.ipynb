```python
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

# Calcul des statistiques descriptives de base
moyenne = stats.tmean(data)
# Remarque : La médiane se calcule avec NumPy ou la méthode descriptive globale
mode_result = stats.mode(data, keepdims=True)
variance = stats.tvar(data)
ecart_type = stats.tstd(data)

print(f"Moyenne    : {moyenne}")
print(f"Mode       : {mode_result.mode[0]} (apparaît {mode_result.count[0]} fois)")
print(f"Variance   : {variance:.2f}")
print(f"Écart-type : {ecart_type:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Paramètres de la distribution normale
moyenne_dist = 50
ecart_type_dist = 10

# Génération de 1000 points entre 20 et 80 pour tracer la courbe propre
x = np.linspace(20, 80, 1000)
# Calcul de la Densité de Probabilité (PDF)
y = norm.pdf(x, loc=moyenne_dist, scale=ecart_type_dist)

# Graphique
plt.figure(figsize=(8, 4))
plt.plot(x, y, color='#0D9488', linewidth=2.5, label='Distribution Normale')
plt.fill_between(x, y, color='#0D9488', alpha=0.1) # Remplissage turquoise
plt.title('Distribution Normale ($\mu=50, \sigma=10$)', fontweight='bold')
plt.xlabel('Valeurs')
plt.ylabel('Densité de probabilité')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
from scipy import stats

# Fixer la graine pour la reproductibilité
np.random.seed(42)

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Exécution du test t pour deux échantillons indépendants
stat_t, p_val = stats.ttest_ind(data1, data2)

print("=== TEST T INDÉPENDANT ===")
print(f"Statistique t : {stat_t:.4f}")
print(f"p-value       : {p_val:.4e}")

if p_val < 0.05:
    print("👉 Conclusion : Les moyennes des deux groupes sont significativement DIFFÉRENTES.")
else:
    print("👉 Conclusion : Pas de différence significative entre les deux groupes.")
```

```python
import numpy as np
from scipy import stats

house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]

# Calcul de la régression linéaire
pente, ordonnee_origine, r_val, p_val, std_err = stats.linregress(house_sizes, house_prices)

# Réactions aux questions du cours :
# 1. Pente et ordonnée à l'origine
print(f"1. Pente : {pente:.2f} | Ordonnée à l'origine : {ordonnee_origine:.2f}")

# 2. Estimation pour 90 m²
taille_test = 90
prix_estime = (pente * taille_test) + ordonnee_origine
print(f"2. Prix estimé pour une maison de 90 m² : {prix_estime:,.2f} unités monétaires")

# 3. Interprétation
print(f"3. Interprétation de la pente : Chaque mètre carré supplémentaire augmente le prix de la maison de {pente:.2f} unités monétaires.")
```

```python
from scipy import stats

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Exécution de l'ANOVA
valeur_f, p_val = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("=== TEST ANOVA : ENGRAIS ===")
print(f"Valeur F : {valeur_f:.4f}")
print(f"Valeur p : {p_val:.4e}")

# Réponses aux questions :
print("\nRéponses aux questions :")
if p_val < 0.05:
    print("- Oui, les engrais ont un effet significativement différent sur la croissance des plantes.")
else:
    print("- Non, aucune différence significative constatée.")

print("- Si la valeur P était supérieure à 0.05 : Nous aurions conclu que les différences constatées entre les trois groupes de plantes sont dues au pur hasard et que les trois engrais ont en réalité la même efficacité.")
```

```python
from scipy.stats import binom

# n = nombre de lancers (10), p = probabilité de succès (0.5 pour face), k = succès souhaités (5)
n = 10
p = 0.5
k = 5

probabilite = binom.pmf(k, n, p)
print(f"Probabilité d'obtenir exactement 5 faces en 10 lancers : {probabilite * 100:.2f} %")
```

```python
import pandas as pd
from scipy import stats

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Corrélation de Pearson (linéaire)
pearson_coef, pearson_p = stats.pearsonr(data['age'], data['income'])

# Corrélation de Spearman (rangs / non-linéaire)
spearman_coef, spearman_p = stats.spearmanr(data['age'], data['income'])

print(f"Coefficient de Pearson  : {pearson_coef:.4f} (p-value: {pearson_p:.4f})")
print(f"Coefficient de Spearman : {spearman_coef:.4f} (p-value: {spearman_p:.4f})")
print("👉 Note : Un coefficient de 1.0 indique une corrélation positive parfaite (le revenu augmente de façon strictement prévisible avec l'âge).")
```
