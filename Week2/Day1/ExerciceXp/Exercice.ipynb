# 📊 Exercice 1 — Introduction à l'Analyse des Données

---

## 1. Qu'est-ce que l'analyse de données ?

L'analyse de données est le processus d'inspection, de nettoyage, de transformation 
et de modélisation des données dans le but de découvrir des informations utiles et 
d'aider à la prise de décision.

Les étapes clés sont :
1. **Collecte** — rassembler les données depuis diverses sources
2. **Nettoyage** — corriger les erreurs et valeurs manquantes
3. **Exploration** — comprendre la structure et les distributions
4. **Modélisation** — appliquer des méthodes statistiques ou ML
5. **Communication** — présenter les résultats de façon claire

---

## 2. Pourquoi l'analyse de données est-elle importante ?

- 📌 **Prise de décision éclairée** — remplacer les suppositions par des faits
- 📌 **Optimisation des processus** — identifier les inefficacités
- 📌 **Anticipation des tendances** — détecter les évolutions du marché
- 📌 **Innovation produit** — mieux comprendre les besoins des clients
- 📌 **Avantage concurrentiel** — surpasser les concurrents grâce aux données

---

## 3. Trois domaines d'application

### 🏥 Santé et Médecine
Utilisation de l'analyse de données pour la médecine prédictive et personnalisée.  
*Exemple : détection précoce du sepsis chez les patients hospitalisés.*

### 🛒 Commerce et Marketing
Systèmes de recommandation basés sur l'historique d'achat et le comportement client.  
*Exemple : Amazon, Jumia personnalisent l'expérience de chaque utilisateur.*

### 🌾 Agriculture de Précision
Capteurs IoT + analyse de données pour optimiser l'irrigation et anticiper les maladies.  
*Exemple : startups africaines comme Hello Tractor modernisant le secteur agricole.*

---

## Conclusion

L'analyse de données est une compétence fondamentale dans presque tous les secteurs. 
Comprendre ses principes est la première étape pour devenir un professionnel 
orienté données dans le monde moderne.

# 📊 Exercice 3 — Identification des Types de Données

Pour chaque dataset, on catégorise chaque colonne comme :
- **Quantitative** : valeur numérique mesurable (âge, salaire, durée...)
- **Qualitative** : catégorie ou étiquette (sexe, pays, statut...)

```python
print("=== Dataset 1 : Sleep Americans ===")
print(df_sleep.dtypes)

print("\n=== Dataset 2 : Mental Health ===")
print(df_mental.dtypes)

print("\n=== Dataset 3 : Credit Card ===")
print(df_credit.dtypes)
```

```python
print("=== Classification Dataset 1 : Sleep ===")
for col in df_sleep.columns:
    if df_sleep[col].dtype in ['int64', 'float64']:
        print(f"  {col} → Quantitative")
    else:
        print(f"  {col} → Qualitative")

print("\n=== Classification Dataset 2 : Mental Health ===")
for col in df_mental.columns:
    if df_mental[col].dtype in ['int64', 'float64']:
        print(f"  {col} → Quantitative")
    else:
        print(f"  {col} → Qualitative")

print("\n=== Classification Dataset 3 : Credit Card ===")
for col in df_credit.columns:
    if df_credit[col].dtype in ['int64', 'float64']:
        print(f"  {col} → Quantitative")
    else:
        print(f"  {col} → Qualitative")
```

# 📊 Exercice 6 — Identification des Types de Données (Structurées vs Non Structurées)

| Source | Type | Justification |
|---|---|---|
| Rapports financiers dans un fichier Excel | ✅ Structurée | Organisé en lignes/colonnes avec un schéma défini, interrogeable directement |
| Photographies sur les réseaux sociaux | ❌ Non Structurée | Fichiers binaires sans organisation interne, nécessite la vision par ordinateur |
| Collection d'articles de presse | ❌ Non Structurée | Texte libre sans schéma rigide, nécessite le NLP pour en extraire de l'info |
| Données d'inventaire en base de données | ✅ Structurée | Tables avec colonnes typées, relations formalisées, interrogeable en SQL |
| Enregistrements audio d'entretiens | ❌ Non Structurée | Fichiers audio sans organisation tabulaire, nécessite transcription + NLP |
